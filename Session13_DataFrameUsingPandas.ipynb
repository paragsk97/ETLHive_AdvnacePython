- loc - label
- iloc - index location

```python
#Using iloc method
df.iloc[0,0]
```

```python
#Extract Painting
df
```

```python
df.loc['S4','Hobby']
```

```python
df.iloc[3,4]
```

```python
df.loc[['S4','S5'], ['Name','Hobby']]
```

```python
df.iloc[[3,4],[0,4]]
```
