## Create function to add two numbers 

```python
def add(x:int|float , y : int|float):
    """
    add two numbers
    """
    z = x + y
    return z
```

```python
add(x = 34, y = 36)
```

```python
def add1(x:int|float , y : int|float):
    """
    add two numbers
    """
    if isinstance(x, (int, float)) and (y, (int, float)):
        z = x + y
        return z
    else:
        print("Data should be numeric")
```

```python
add1(12,4)
```

```python
add1('data','science')
```

```python
print(add1.__doc__)
```

## use assert

```python
def add2(x:int|float , y : int|float):
    """
    add two numbers
    """
    assert isinstance(x, (int, float)) , 'x should be numeric'  
    assert isinstance(y, (int, float)) , 'y should be numeric'
    z = x + y
    return z
```

```python
add2('data','science')
```

## Variable length arguments --> *args

### create function to add n arguments

```python
def add3(*args):
    print(args)
    print(type(args))
```

```python
add3(1,2,3)
```

```python
def add_args(*args):
    add = 0
    for i in args:
        add = add + i
    return add
```

```python
add_args(1,-5,6,6473,234445)
```

```python
add_args(-3245, 456, 8, 78, 24)
```

```python
add_args(1.5, 3, 90, 55, 44, 30, 67)
```

```python
## create function for bill with tax
def total_bill(*args):
    price_after_tax = 0
    bill_before_tax = 0
    for product , price , quantity in args:
        subtotal = price*quantity
        print(f'subtotal of {product} --> {subtotal}')

        bill_before_tax = bill_before_tax + subtotal

        print(f"Bill before tax --> {bill_before_tax}")

        price_after_tax = bill_before_tax + bill_before_tax * 0.18

    return f"Bill after tax --> {price_after_tax}"
```

```python
total_bill(('mobile', 25000, 2))
```

```python
total_bill(('mobile', 25000, 2), ('Laptop', 65000, 3))
```

```python
def total_bill2(*args):
    bill_before_tax = 0
    price_before_tax = 0
    for product, price, quantity, discount in args:
        subtotal = price * quantity
        discount_amount = subtotal * (discount / 100)
        subtotal_after_discount = subtotal - discount_amount
        print(f'subtotal of {product} --> {subtotal}')
        bill_before_tax += subtotal_after_discount
        print(f'Discount of {product} --> {discount}')
        print(f'Discount of {product} after discount --> {subtotal_after_discount}')

    print(f"Bill before tax --> {bill_before_tax}")

    price_after_tax = bill_before_tax + (bill_before_tax * 0.18)
    print(f"Bill after tax --> {price_after_tax}")

    return price_after_tax
```

```python
total_bill2(('mobile', 25000, 2, 10), ('Laptop', 65000, 3, 5))
```

```python
## Other way
def  total_bill_with_discount(*args):
    bill_before_tax =  0
    price_after_tax  =  0
    for product  , price  ,  quantity  , discount in  args:
        subtotal  =  price*quantity
        bill_with_discount  =  subtotal - subtotal*discount/100
        print(f'subtotal  of {product} --> {subtotal}')
        print(f"bill with discount --> {bill_with_discount }")
        bill_before_tax  =  bill_before_tax + bill_with_discount
        price_after_tax  =  bill_before_tax + bill_before_tax*0.18
    print(f"Bill before tax  --> {bill_before_tax}")
    return f"bill after tax --> {price_after_tax}"
```
