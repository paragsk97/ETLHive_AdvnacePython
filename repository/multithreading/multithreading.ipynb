```python
# create function to find hypotenus
def Hypotenus(a : int|float , b:int|float):
    print("Hypotenus function is started")
    time.sleep(2)
    h = (a**2 + b**2)*(1/2)
    print("Hypotenus function is completed")
    print(f"Hypotenus is  --> {h}")
   
```

```python
%%time
Hypotenus(a = 3 , b = 5)
```

```python
%%time
Hypotenus(3, 7)
Hypotenus(7, 12)
```

```python
# create function for finding simple interest
def simple_interest(P, N, R):
    print("Simple interest function is started")
    time.sleep(3)
    I  = P*N*R/100
    A = P+I
    print("Simple interest function is completed")
    return f"P : {P}, N : {N} , R : {R} , I : {I} , A : {A}"
    
```

```python
%%time
simple_interest(P = 100000 , N = 2 , R = 6.5)
```

## create threads for simultaneous working 

```python
from threading import Thread
```

```python
%%time
th1 = Thread(target = Hypotenus , args = ( 3 , 6))
th2 = Thread(target = simple_interest , args = ( 10000 , 2 , 5.5))

#start the threads
th1.start()
th2.start()

#join the threads
th1.join()
th2.join()
```

```python
#create function for factorial of a number 
def factorial(n):
    print("Factorial function is started")
    fact  = 1
    for i in range(1, n + 1):
        fact *= i
    print(f"Factorial of {n} is : {fact}")
    print("Factorial function is completed")
```

```python
%%time
factorial(6)
```

```python
l = [12, 12, 15, 3, 4]
```

```python
%%time
for i in l:
    factorial(i)
```

```python
#create function multithreading
def multithread(l):
    threads = []
    for i in l:
        th = Thread(target = factorial , args = (i,))
        th.start()
        threads.append(th)

#end the threads
    for th in threads:
        th.join()
```

```python
%%time 
multithread([12,4,6,8,19])
```

```python
from urllib.request import urlretrieve
```

```python
url = 'https://raw.githubusercontent.com/utkarshg1/mlproject_regression/main/artifacts/data.csv'
```

```python
url.split('/')
```

```python
#create function to download the file
def file_download(url):
    file_name = url.split('/')[-1]
    print(f"{file_name} is downloading")
    urlretrieve(url , file_name)
    print(f"{file_name} is downloaded successfully")
```

```python
file_download(url)
```

## create multi thread function for downloading multiple files

```python
def down_multi_files(l):
     threads = []
     for i in l:
        th = Thread(target = file_download , args = (i,))
        th.start()
        threads.append(th)

#end the threads
     for th in threads:
        th.join()
```
