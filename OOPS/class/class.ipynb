### dataclass for defining attributes

```python
from dataclasses import dataclass
from typing import Literal , List
```

### Create class Student
- Name, Age, Marks, Hobbies, Gender

```python
@dataclass
class Student:
    Name : str
    Age : int
    Marks : float
    Gender : Literal["Male", "Female"]
    Hobbies : List[str]

    def get_student_info(self):
        print(f"I am {self.Name}")
        print(f"I am {self.Age} years old")
        print(f"Marks I got {self.Marks}")
        print(f"I am {self.Gender}")
        print(f"My Hobbies are {self.Hobbies}")
```

### Create class for bank account

```python
@dataclass
class Account:
    acc_name : str
    acc_no : int
    bal : float

    #Create method to check balance

    def check_balance(self):
        print(f"Current Balance Account Number of {self.acc_no} is {self.bal}")

    # Deposit Amount
    def Deposit(self, Amt):
        self.bal = self.bal + Amt
        print(f"Amount of {Amt} is Deposited in Account {self.acc_no}")  # type: ignore
        return self.check_balance()
    
    # Withdraw Amount
    def Withdraw(self, Amt : float):
        if Amt > self.bal:
            raise ValueError("Insufficient Balance")
        else:
            self.bal = self.bal - Amt
            print(f"Amount of {Amt} is Withdrawed in Account {self.acc_no}")  # type: ignore
            return self.check_balance()
        
    # Transfer Amount from one account of another account
    def Transfer(self, Amt : float, Acc = "Account"):
        self.Withdraw(Amt=Amt)
        Acc.Deposit(Amt) # type: ignore
```

```python
a1 = Account(acc_name = "Sarthak" , acc_no= 123 , bal= 2000)
```

```python
a1.acc_no
```

```python
a1.check_balance()
```
