```python
from dataclasses import dataclass 
```

```python
@dataclass
class Triangle():
    a : int|float
    b : int|float
    c : int|float

    def Perimeter(self):
        if(self.a + self.b > self.c) and \
            (self.b + self.c > self.a) and  \
           (self.a + self.c > self.b):
            p = self.a + self.b + self.c
            return p
        
    def Area(self):
        if self.Perimeter() is not None:
            s  =  self.Perimeter()/2 # type: ignore
            Area  =  (s*(s-self.a)*(s-self.b)*(s-self.c))**1/2
            return f"Area of Triangle is {Area}"
        else:
            print(f"Sides do not form Triangle")
```

```python
t1 = Triangle(a = 2 , b = 41 , c = 14)
```

```python
t1.Perimeter()
```

```python
print(t1.Perimeter())
```

```python
t1.Area()
```

```python
print(t1.Area())
```

```python
t2 = Triangle(a = 13, b = 14, c = 6)
```

```python
t2.Perimeter()
```

```python
t2.Area()
```

```python
t3 = Triangle(a = 34, b = 13, c = 23)
```

```python
for tr in [t1, t2, t3]:
    print(tr) 
    print("Perimeter --> ", tr.Perimeter() )
    print("Area -- >", tr.Area())
    print("--------------------------------")
```

### Create Simple Class for creating Calculator 
- addition
- substraction
- multiplication
- division

```python
class Simple_Calculator:
 
    def __init__(self, a, b):
        self.a = a
        self.b = b
 
    def add(self):
        if isinstance(self.a  ,(int , float)) and  isinstance(self.b  ,(int , float)):
            return self.a + self.b
 
    def subtract(self):
        if self.add() is not None:
            return self.a - self.b
 
    def multiply(self):
        if self.add() is not None:
            return self.a * self.b
 
    def divide(self):
        if self.add() is not None:
            if self.b !=0 :
                return self.a / self.b
            else:
                raise ZeroDivisionError("Denominator Should not be zero")
```

```python
o1 = Simple_Calculator(a = 23, b = 4)
```

```python
o1.add()
```

```python
o1.subtract()
```

```python
o1.multiply()
```

```python
o1.divide()
```

## Inheritance

```python
# Create class person and employee
class Person :
    def __init__(self, name, age):
        self.name = name
        self.age  = age
    
    def introduce(self):
        return f"I am {self.name} and I am {self.age} years old"


#create employee class
class employee(Person):
    def __init__(self, name, age, company):
        super().__init__(name, age)
        self.company = company
        
    def work(self):
        print(f" {super().introduce()} , I work for {self.company}")
```

```python
p1 = Person(name = 'Sunita' , age= 23)
```

```python
p1.name
```

```python
p1.age
```

```python
e1 = employee(name = 'Suresh', age = 23, company = "TCS")
```

```python
e1.introduce()
```

```python
e1.work()
```

## Multilevel Inheritance

```python
# Create class for employee, manager, project manager
class Employee:
    def __init__(self, name, emp_id):
        self.name = name
        self.emp_id = emp_id

    def get_employee_info(self):
        return f"Name : {self.name} , ID : {self.emp_id}"
    
# create another class manager
class Manager(Employee):
    def __init__(self, name, emp_id, department):
        super().__init__(name, emp_id)
        self.department = department

    #create method
    def get_manager_info(self):
        return f"{self.get_employee_info()}, Department : {self.department}"
    
#create Project Manager class
class Project_Manager(Manager):
    def __init__(self, name, emp_id, department, Project):
        super().__init__(name, emp_id, department)
        self.Project = Project

    #create method   
    def get_project_manager_info(self):
        return f"{self.get_manager_info()}, Project : {self.Project}" 
```

```python
e1 = Employee(name = "Surendra" , emp_id= 123)
```

```python
e1.name
```

```python
e1.emp_id
```

```python
e1.get_employee_info()
```

```python
m1 = Manager(name="Sushmita", emp_id=234, department="IT")
```

```python
m1.get_manager_info()
```

```python
m2 = Project_Manager(name="Sushmita", emp_id=234, department="IT", Project="Mobile Application")
```

```python
m2.get_project_manager_info()
```
